Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
# Групування даних за стовпцем "Посада" та підрахунок кількості записів у кожній групі
grouped = df.groupby('Посада').size()

# Перегляд результатів
print(grouped)
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата на місяць"

```python
# Групування за стовпцем "Посада" і агрегація даних
aggregated = df.groupby('Посада').agg({
    'Зарплата на місяць': ['min', 'max']  # Застосування функцій min і max до стовпця "Зарплата.в.місяць"
})

# Перегляд результатів
print(aggregated)
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
# Функція для обчислення середньої заробітної плати на місяць
def fill_avg_salary(series):
    return series.mean()

# Додавання нового стовпчика "avg" з середнім значенням зарплати по всьому DataFrame
df['avg'] = df['Зарплата на місяць'].apply(lambda x: fill_avg_salary(df['Зарплата на місяць']))

# Перегляд перших кількох рядків для перевірки
print(df.head())

# Перевірка нових стовпчиків
print(df[['Зарплата на місяць', 'avg']].head())
```

```python
# Описова статистика для нового стовпчика "avg"
stats = df['avg'].describe()

# Перегляд описової статистики
print(stats)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from sklearn.preprocessing import LabelEncoder

# Читання CSV-файлу
df = pd.read_csv('/content/2017_jun_final.csv')

# Закодуємо рівень англійської числовими значеннями
le = LabelEncoder()
df['Рівень англійської кодування'] = le.fit_transform(df['Рівень англійської'])

# Витягування значень для осей
X = df['Рівень англійської кодування']
Y = df['Вік']
Z = df['Зарплата на місяць']

# Створення сітки для X і Y
X_grid, Y_grid = np.meshgrid(np.unique(X), np.unique(Y))

# Інтерполяція Z значень на сітку
Z_grid = np.zeros_like(X_grid, dtype=float)
for i in range(len(np.unique(X))):
    for j in range(len(np.unique(Y))):
        mask = (X == X_grid[j, i]) & (Y == Y_grid[j, i])
        if np.any(mask):
            Z_grid[j, i] = np.mean(Z[mask])

# Створення тривимірного графіка
fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111, projection='3d')

# Побудова поверхні
surf = ax.plot_surface(X_grid, Y_grid, Z_grid, cmap=cm.viridis, edgecolor='k')

# Додавання кольорової шкали
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# Налаштування осей
ax.set_title('Залежність зарплати від рівня англійської та віку', fontsize=16)
ax.set_xlabel('Рівень англійської (кодування)', fontsize=14)
ax.set_ylabel('Вік', fontsize=14)
ax.set_zlabel('Зарплата (за місяць)', fontsize=14)

# Показати графік
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Читання CSV-файлу
df = pd.read_csv('/content/2017_jun_final.csv')

# Вибір міст, що нас цікавлять
cities_of_interest = ['Київ', 'Харків', 'Львів', 'Івано-Франківськ', 'Дніпро', 'Одеса']
df_cities = df[df['Місто'].isin(cities_of_interest)]

# Створення графіків для кожного міста
plt.figure(figsize=(16, 12))

for i, city in enumerate(cities_of_interest, 1):
    df_city = df_cities[df_cities['Місто'] == city].sort_values(by='Загальний досвід роботи')

    plt.subplot(2, 3, i)
    sns.scatterplot(data=df_city,
                    x='Зарплата на місяць',
                    y='Загальний досвід роботи',
                    alpha=0.7,
                    color='blue')
    plt.title(f'{city}', fontsize=14)
    plt.xlabel('Зарплата (в місяць)', fontsize=12)
    plt.ylabel('Загальний досвід роботи (роки)', fontsize=12)

plt.tight_layout()
plt.show()
```
