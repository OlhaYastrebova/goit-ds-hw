Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
# Порахуємо кількість пропущених значень у кожному стовпці
missing_values = df.isnull().sum()

# Обчислимо загальну кількість рядків у DataFrame
total_rows = len(df)

# Розрахуємо частку пропусків у кожному стовпці
missing_percentage = (missing_values / total_rows) * 100

# Виведемо частку пропусків
print(missing_percentage)
```

Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

```python
# Стовпець, який не потрібно видаляти
column_to_keep = 'Мова програмування'

# Визначення стовпців з пропусками
columns_with_missing = df.columns[df.isnull().any()]

# Видаляємо стовпці з пропусками, крім вказаного стовпця
columns_to_drop = [col for col in columns_with_missing if col != column_to_keep]

# Видалення стовпців
df_cleaned = df.drop(columns=columns_to_drop)

# Перегляд перших кількох рядків для перевірки
print(df_cleaned.head())

# Перевірка розміру після видалення стовпців
print(f'Новий розмір таблиці: {df_cleaned.shape[0]} рядків і {df_cleaned.shape[1]} стовпців')
```

```python
# Перевірка, чи залишився тільки стовпець "Мова програмування"
print(f'Стовпці після видалення: {df_cleaned.columns.tolist()}')

# Перевірка частки пропусків у кожному стовпці
missing_values = df_cleaned.isnull().sum()
total_rows = len(df_cleaned)
missing_percentage = (missing_values / total_rows) * 100

# Виведення частки пропусків
print('Частка пропусків у кожному стовпці:')
print(missing_percentage)
```

Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
# Групування даних за стовпцем "Посада" та підрахунок кількості записів у кожній групі
grouped = df.groupby('Посада').size()

# Перегляд результатів
print(grouped)
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата на місяць"

```python
# Групування за стовпцем "Посада" і агрегація даних
aggregated = df.groupby('Посада').agg({
    'Зарплата на місяць': ['min', 'max']  # Застосування функцій min і max до стовпця "Зарплата.в.місяць"
})

# Перегляд результатів
print(aggregated)
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
# Функція для обчислення середньої заробітної плати на місяць
def fill_avg_salary(series):
    return series.mean()

# Додавання нового стовпчика "avg" з середнім значенням зарплати по всьому DataFrame
df['avg'] = df['Зарплата на місяць'].apply(lambda x: fill_avg_salary(df['Зарплата на місяць']))

# Перегляд перших кількох рядків для перевірки
print(df.head())

# Перевірка нових стовпчиків
print(df[['Зарплата на місяць', 'avg']].head())
```

Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
# Описова статистика для нового стовпчика "avg"
stats = df['avg'].describe()

# Перегляд описової статистики
print(stats)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Читання CSV-файлу
df = pd.read_csv('/content/2017_jun_final.csv')

# Створення середньої зарплати для кожної посади
average_salaries = df.groupby('Посада')['Зарплата на місяць'].mean().reset_index()

# Налаштування стилю графіка
sns.set(style="whitegrid")

# Створення графіка
plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(x='Посада', y='Зарплата на місяць', data=average_salaries, color='teal')

# Налаштування заголовка та підписів
plt.title('Середня зарплата різних посад', fontsize=16)
plt.xlabel('Посада', fontsize=14)
plt.ylabel('Середня зарплата (за місяць)', fontsize=14)
plt.xticks(rotation=45, ha='right')

# Показати графік
plt.tight_layout()
plt.show()

```
