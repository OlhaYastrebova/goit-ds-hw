напишіть функцію гіпотези лінійної регресії у векторному вигляді

```python
import numpy as np
import pandas as pd

def hypothesis(X, w):
    return np.dot(X, w)
```

створіть функцію для обчислення функції втрат у векторному вигляді

```python
def compute_cost(X, y, w):
    m = len(y)  # Кількість навчальних прикладів
    predictions = np.dot(X, w)  # Обчислення гіпотези
    errors = predictions - y  # Відхилення від реальних значень
    cost = (1 / (2 * m)) * np.dot(errors.T, errors)  # Функція втрат (середньоквадратична помилка)
    return cost
```

реалізуйте один крок градієнтного спуску

```python
def gradient_descent_step(X, y, w, alpha):
    m = len(y)  # Кількість навчальних прикладів

    # Обчислення передбачень моделі
    predictions = np.dot(X, w)

    # Обчислення похибок (відхилень від реальних значень)
    errors = predictions - y

    # Обчислення градієнтів
    gradient = (1 / m) * np.dot(X.T, errors)

    # Оновлення параметрів
    w = w - alpha * gradient

    return w
```

знайдіть найкращі параметри *w*
  для датасету використовуючи написані вами функції, прогнозуючу ціну на будинок залежно від площі, кількості ванних кімнат та кількості спалень

```python
# Прочитайте CSV файл
df = pd.read_csv('/content/Housing.csv')
df
```

```python
# Вибір ознак для X
X = df[['area', 'bathrooms', 'bedrooms']].values

# Вибір цільової змінної для y
y = df['price'].values

# Додавання стовпця одиниць для w_0 (зсув)
X = np.hstack((np.ones((X.shape[0], 1)), X))

print("X:", X)
print("y:", y)
```

```python
# Ініціалізуємо параметри (початково всі нулі)
w = np.zeros(X.shape[1])

# Кількість ітерацій
iterations = 1000

# Коефіцієнт навчання (learning rate)
alpha = 0.01

def compute_cost(X, y, w):
    m = len(y)
    predictions = np.dot(X, w)
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return cost

def gradient_descent(X, y, w, alpha, iterations):
    m = len(y)
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        w = gradient_descent_step(X, y, w, alpha)
        cost_history[i] = compute_cost(X, y, w)

    return w, cost_history

# Нормалізація ознак
mean_X = np.mean(X[:, 1:], axis=0)
std_X = np.std(X[:, 1:], axis=0)
X[:, 1:] = (X[:, 1:] - mean_X) / std_X

# Перезапуск градієнтного спуску
optimal_w, cost_history = gradient_descent(X, y, w, alpha, iterations)

print("Оптимальні параметри w:", optimal_w)
```

знайдіть ці ж параметри за допомогою аналітичного рішення

```python
# Використання аналітичного рішення для знаходження параметрів w
def normal_equation(X, y):
    return np.linalg.inv(X.T @ X) @ X.T @ y

# Знаходимо оптимальні параметри w за допомогою нормальних рівнянь
optimal_w_analytical = normal_equation(X, y)

print("Оптимальні параметри w (нормальні рівняння):", optimal_w_analytical)
```

для перевірки спрогнозованих значень, використайте LinearRegression з бібліотеки scikit-learn та порівняйте результати

```python
from sklearn.linear_model import LinearRegression

# Створення та тренування моделі LinearRegression
model = LinearRegression()
model.fit(X[:, 1:], y)  # Виключаємо перший стовпець з одиницями, бо sklearn додає його автоматично

# Отримання прогнозованих параметрів w
w_sklearn = np.hstack((model.intercept_, model.coef_))

print("Оптимальні параметри w (scikit-learn):", w_sklearn)
```

```python
# Прогнозування з використанням параметрів від градієнтного спуску
predictions_gd = X @ optimal_w

# Прогнозування з використанням параметрів від нормальних рівнянь
predictions_ne = X @ optimal_w_analytical

# Прогнозування з використанням scikit-learn
predictions_sklearn = model.predict(X[:, 1:])

# Виведення результатів
print("Прогнози (градієнтний спуск):", predictions_gd)
print("Прогнози (нормальні рівняння):", predictions_ne)
print("Прогнози (scikit-learn):", predictions_sklearn)
```
